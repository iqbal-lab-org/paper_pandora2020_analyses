```python
import pandas as pd
```

```python
# preprocessing
!cp ../pandora1_paper_analysis_output_20_way/illumina_analysis/ROC_data.tsv ROC_data_illumina.tsv
!cp ../pandora1_paper_analysis_output_20_way/illumina_analysis_with_filters/ROC_data.tsv ROC_data_illumina_filters.tsv
!cp ../pandora1_paper_analysis_output_20_way/nanopore_analysis/ROC_data.tsv ROC_data_nanopore.tsv
!cp ../pandora1_paper_analysis_output_20_way/nanopore_analysis_with_filters/ROC_data.tsv ROC_data_nanopore_filters.tsv
```

```python
df_illumina = pd.read_csv("ROC_data_illumina.tsv", sep="\t")
df_illumina_filtered = pd.read_csv("ROC_data_illumina_filters.tsv", sep="\t")
df_nanopore = pd.read_csv("ROC_data_nanopore.tsv", sep="\t")
df_nanopore_filtered = pd.read_csv("ROC_data_nanopore_filters.tsv", sep="\t")

df = pd.concat([df_illumina, df_illumina_filtered, df_nanopore, df_nanopore_filtered], ignore_index=True).drop(columns=['Unnamed: 0'])
df
```

```python
# filtering for step_GT == 0
df = df[df.step_GT == 0]
df
```

```python
# give a category for the tool
def get_tool_category(tool):
    if tool.startswith("pandora"):
        return tool.replace("_global_genotyping", "")
    else:
        return tool.split("_")[0]

df["tool_category"] = df["tool"].apply(get_tool_category)
df
```

```python
# cleanup
df = df[["tool_category", "coverage_threshold", "strand_bias_threshold", "gaps_threshold", "error_rate", "recalls_wrt_variants_found_wrt_alleles", "tool"]]
df.to_csv("df_cleaned_up.csv", index=False)
df
```

```python
# get best precision results - rounded
df_best_precision = \
  df.sort_values(by=["error_rate"])\
  .groupby(by=["tool_category", "coverage_threshold", "strand_bias_threshold", "gaps_threshold"])\
  .head(1)\
  .sort_values(by=["tool_category"])
df_best_precision["error_rate"] = round(df_best_precision["error_rate"]*100, 2)
df_best_precision["recalls_wrt_variants_found_wrt_alleles"] = round(df_best_precision["recalls_wrt_variants_found_wrt_alleles"]*100, 1)
df_best_precision.to_csv("df_best_precision.rounded.csv", index=False)
df_best_precision
```

```python
# get best precision results - not rounded
df_best_precision = \
  df.sort_values(by=["error_rate"])\
  .groupby(by=["tool_category", "coverage_threshold", "strand_bias_threshold", "gaps_threshold"])\
  .head(1)\
  .sort_values(by=["tool_category"])
df_best_precision["error_rate"] = df_best_precision["error_rate"]*100
df_best_precision["recalls_wrt_variants_found_wrt_alleles"] = df_best_precision["recalls_wrt_variants_found_wrt_alleles"]*100
df_best_precision.to_csv("df_best_precision.not_rounded.csv", index=False)
df_best_precision
```

```python
# get best AvgAR results - rounded
df_best_recall = \
  df.sort_values(by=["recalls_wrt_variants_found_wrt_alleles"], ascending=False)\
  .groupby(by=["tool_category", "coverage_threshold", "strand_bias_threshold", "gaps_threshold"])\
  .head(1)\
  .sort_values(by=["tool_category"])
df_best_recall["error_rate"] = round(df_best_recall["error_rate"]*100, 2)
df_best_recall["recalls_wrt_variants_found_wrt_alleles"] = round(df_best_recall["recalls_wrt_variants_found_wrt_alleles"]*100, 1)
df_best_recall.to_csv("df_best_recall.rounded.csv", index=False)
df_best_recall
```

```python
# get best AvgAR results - not rounded
df_best_recall = \
  df.sort_values(by=["recalls_wrt_variants_found_wrt_alleles"], ascending=False)\
  .groupby(by=["tool_category", "coverage_threshold", "strand_bias_threshold", "gaps_threshold"])\
  .head(1)\
  .sort_values(by=["tool_category"])
df_best_recall["error_rate"] = df_best_recall["error_rate"]*100
df_best_recall["recalls_wrt_variants_found_wrt_alleles"] = df_best_recall["recalls_wrt_variants_found_wrt_alleles"]*100
df_best_recall.to_csv("df_best_recall.not_rounded.csv", index=False)
df_best_recall
```
