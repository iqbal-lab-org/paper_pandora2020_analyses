```python
import pandas as pd
import os
```

```python
# read df
four_way_df = pd.read_csv("ROC_data_old_and_new_basecall.tsv", sep="\t")
four_way_df
```

```python
# add new columns
four_way_df["methylation_aware"] = four_way_df["tool"].apply(lambda tool: "yes" if tool.startswith("pandora_NEW_BASECALL") else "no")
four_way_df["discover"] = four_way_df["tool"].apply(lambda tool: "yes" if "withdenovo" in tool else "no")
four_way_df
```

```python
# plot with R
four_way_df.to_csv("ROC_data_old_and_new_basecall.R_data.csv")
os.system("Rscript plot_ROC_4way.R")
os.remove("ROC_data_old_and_new_basecall.R_data.csv")
```
